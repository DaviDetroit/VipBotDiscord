```python
import pandas as pd
import matplotlib.pyplot as plt
import mysql.connector

# Conexão ao banco
conn = mysql.connector.connect(
    host='localhost',
    user='fernandez',
    password='root007',
    database='SistemaVips'
)

# Consulta
query = "SELECT id, nome_discord, data_inicio FROM vips"
df = pd.read_sql(query, conn)
df['data_inicio'] = pd.to_datetime(df['data_inicio'])

# Agrupar por mês/ano
df['mes_ano'] = df['data_inicio'].dt.to_period('M')
contagem = df.groupby('mes_ano').size()

# Plotar gráfico com melhorias visuais
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(9,4))  # mais compacto

bars = ax.bar(contagem.index.astype(str), contagem.values,
              color='#4C72B0', edgecolor='black', linewidth=0.8)

ax.set_title('Cadastros VIP por Mês', fontsize=14, fontweight='bold')
ax.set_xlabel('Mês/Ano', fontsize=11)
ax.set_ylabel('Quantidade', fontsize=11)

ax.set_xticklabels(contagem.index.astype(str), rotation=45, ha='right', fontsize=9)

# Mostrar valor no topo das barras
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # distância do texto para a barra
                textcoords='offset points',
                ha='center', va='bottom',
                fontsize=9,
                color='black')

plt.tight_layout()
plt.show()
```
