```python
import pandas as pd
import matplotlib.pyplot as plt
import mysql.connector

# Configuração da conexão - substitua pela sua senha e banco
conn = mysql.connector.connect(
    host='localhost',
    user='fernandez',
    password='root007',
    database='SistemaVips'
)

# Consulta SQL para pegar os dados
query = "SELECT id, nome_discord, data_inicio FROM vips"

# Lê os dados direto para o pandas
df = pd.read_sql(query, conn)

# Converte para datetime
df['data_inicio'] = pd.to_datetime(df['data_inicio'])

# Agrupa por mês/ano
df['mes_ano'] = df['data_inicio'].dt.to_period('M')

# Conta quantos cadastros por mês
contagem = df.groupby('mes_ano').size()

# Plotando o gráfico
plt.figure(figsize=(10,5))
contagem.plot(kind='bar', color='skyblue')
plt.title('Cadastros VIP por Mês')
plt.xlabel('Mês/Ano')
plt.ylabel('Quantidade de Cadastros')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
